## Plotting: pyplot submodule

```python
import matplotlib.pyplot as plt
import numpy as np
```

### Plot graphs 


#### First example
Marker properties, linestyle, title and labels properties, font properties, and ticks

```python
# Font settings
xpoints = [0, 1, 2, 3, 4]
ypoints = [0, 4, 8, 6, 8]

# Plotting graph
## Setting marker type (marker), marker size (ms), market edge color (mec) and marker face color (mfc)
## Setting line style (linestyle), line color (c), and line width (lw)
plt.plot(xpoints, ypoints, marker = "o", ms = "8", mec = "b", mfc = "w", linestyle = "--", c = "#4CAF50", lw = "2")
# You can also do it with format strings (fmt)
# For example: plt.plot(xpoints, ypoints, "o:g")

# Creating title and labels
## Setting font properties (fontdict)
## Title positioning (loc)
plt.title("First Graph", fontdict={"fontname": "Times New Roman", "fontsize": 20}, loc="left")
plt.xlabel("X axis")
plt.ylabel("Y axis")

# Ticks
plt.xticks([0, 1, 2, 3, 4])
plt.yticks([0, 2, 4, 6, 8])

# Adding grid lines to the plot
plt.grid()

# Show graph
plt.show()
```

#### Third example
Subplots, supertitles

```python
import matplotlib.pyplot as plt
import numpy as np

# Display multiple plots
## The subplot() function  takes three arguments that describes the layout of the figure. The layout is organized in rows and columns, which are represented by the first and second argument. The third argument represents the index of the current plot. 

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

# the figure has 1 row, 2 columns, and is the first plot
plt.subplot(1, 2, 1)
plt.plot(x,y)
plt.title("First part", loc = "right")

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

# the figure has 1 row, 2 columns, and is the second plot
plt.subplot(1, 2, 2)
plt.plot(x,y, ls = "--")
plt.title("Second part", loc = "right")

# Adding a custom title for the entire figure
supfont = {'family':'serif','color':'blue','size':20}
plt.suptitle("Third graph", fontdict= supfont)
plt.show()
```

### Scatter plot graphs

#### First example
Basic customization

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x, y, c = "#88c999")

plt.xticks([])
plt.yticks([70, 75, 80, 85, 90, 95, 100, 105, 110, 115, 120])
plt.show()
```

#### Second example
Scatter plot with two sets


```python
#day one, the age and speed of 13 cars:
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y)

#day two, the age and speed of 15 cars:
x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y)

# The two plots are plotted with tho different colours, by default blue and orange.
plt.show()
```

#### Third example
Marker size depending on the value of y axis

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,500,86])

# Setting a specific color for each dot. 
## When doing this, you can only use the c argument
colors = np.array(["red","green","blue","yellow","pink","black","orange","purple","beige","brown","gray","cyan","magenta"])

plt.scatter(x, y, c=colors, s = y)

plt.show()
```
