### Plot graphs 


#### First example
Marker properties, linestyle, title and labels properties, font properties, and ticks

```python
# Font settings
xpoints = [0, 1, 2, 3, 4]
ypoints = [0, 4, 8, 6, 8]

# Plotting graph
## Setting marker type (marker), marker size (ms), market edge color (mec) and marker face color (mfc)
## Setting line style (linestyle), line color (c), and line width (lw)
plt.plot(xpoints, ypoints, marker = "o", ms = "8", mec = "b", mfc = "w", linestyle = "--", c = "#4CAF50", lw = "2")
# You can also do it with format strings (fmt)
# For example: plt.plot(xpoints, ypoints, "o:g")

# Creating title and labels
## Setting font properties (fontdict)
## Title positioning (loc)
plt.title("First Graph", fontdict={"fontname": "Times New Roman", "fontsize": 20}, loc="left")
plt.xlabel("X axis")
plt.ylabel("Y axis")

# Ticks
plt.xticks([0, 1, 2, 3, 4])
plt.yticks([0, 2, 4, 6, 8])

# Adding grid lines to the plot
plt.grid()

# Show graph
plt.show()
```

#### Second example
Plotting without lines, font properties, grid properties

```python
# Font settings
x1points = [0, 1, 2, 3, 4]
y1points = [8, 3, 8, 6, 8]
x2points = [0, 3, 5, 4 ,4]
y2points = [1, 3, 4, 5, 6]

# Plotting two graphs without line, only markers
plt.plot(xpoints, "o", marker = "d")
plt.plot(ypoints, "o", marker = "o")

# Setting font properties for title and labels
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

# Setting title and labels
plt.title("Sports Watch Data", fontdict = font1, loc = "right")
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)

# Adding grid lines to the plot
## Specifying grid lines to display (axis)
## Setting line properties for the grid (color; linestyle; linewidht)
plt.grid(axis = "y", color="green", linestyle = "--", linewidth = 0.5)

plt.show()
```

#### Third example
Subplots, supertitles

```python
import matplotlib.pyplot as plt
import numpy as np

# Display multiple plots
## The subplot() function  takes three arguments that describes the layout of the figure. The layout is organized in rows and columns, which are represented by the first and second argument. The third argument represents the index of the current plot. 

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

# the figure has 1 row, 2 columns, and is the first plot
plt.subplot(1, 2, 1)
plt.plot(x,y)
plt.title("First part", loc = "right")

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

# the figure has 1 row, 2 columns, and is the second plot
plt.subplot(1, 2, 2)
plt.plot(x,y, ls = "--")
plt.title("Second part", loc = "right")

# Adding a custom title for the entire figure
supfont = {'family':'serif','color':'blue','size':20}
plt.suptitle("Third graph", fontdict= supfont)
plt.show()
```

### Pie charts

#### First example
Basic customization

```python
# Pie chart
labels = ['Frogs', 'Hogs', 'Dogs', 'Logs']
sizes = [15, 30, 45, 10]

# only "explode" the 2nd slice (i.e. 'Hogs')
explode = (0, 0.1, 0, 0)  

#add colors
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%',
        shadow=True, startangle=90)

# Equal aspect ratio ensures that pie is drawn as a circle
ax1.axis('equal')  

plt.tight_layout()
plt.show()
```

#### Second example
Text customization

```python
fig1, ax1 = plt.subplots()

patches, texts, autotexts = ax1.pie(sizes, colors = colors, labels=labels, autopct='%1.1f%%', startangle=90)

for text in texts:
    text.set_color('grey')

for autotext in autotexts:
    autotext.set_color('black')

# Equal aspect ratio ensures that pie is drawn as a circle
ax1.axis('equal')  
plt.tight_layout()
plt.show()
```

#### Third example
Donut chart

```python
fig1, ax1 = plt.subplots()

ax1.pie(sizes, colors = colors, labels=labels, autopct='%1.1f%%', startangle=90)

# draw circle
centre_circle = plt.Circle((0,0),0.80,fc='white')
fig = plt.gcf() # ?
fig.gca().add_artist(centre_circle) # ?

ax1.axis("equal")
plt.tight_layout()
plt.show()
```

#### Fourth example
Donut chart with label customization and explode

```python
#explsion
explode = (0.05,0.05,0.05,0.05)

fig1, ax1 = plt.subplots()
ax1.pie(sizes, colors = colors, labels=labels, autopct='%1.1f%%', startangle=90, 
        pctdistance=0.83,
        explode = explode)

#draw circle
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

# Equal aspect ratio ensures that pie is drawn as a circle
ax1.axis('equal')  
plt.tight_layout()
plt.show()
```

## Playground: Canadian Inmigration Dataset

### Data preparation

```python
import numpy as np
import pandas as pd

df = pd.read_excel('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DV0101EN/labs/Data_Files/Canada.xlsx',
                       sheet_name='Canada by Citizenship',
                       skiprows=range(20), # just text
                       skipfooter=2) # just text
df.head()
```

```python
df.drop(["AREA", "REG", "DEV", "Type", "Coverage"], axis=1, inplace=True)
df.head()
```

```python
df.rename(columns={"OdName":"Country", "AreaName":"Continent", "RegName":"Region"}, inplace=True)
df.columns
```

```python
df["Total"] = df.sum(axis=1)
df
```

```python
df.isnull().sum()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
plt.style.available
```

```python
mpl.style.use(["ggplot"])
```

```python
# Line plot
# Switzerland's immigration tend to Canada by year
years = list(map(int, range(1980, 2014)))
```

```python
switzerland = df.loc["Switzerland", years]
```

### Line plot

```python
switzerland.plot()
plt.title("Inmigration from Switzerland")
plt.xlabel("Years")
plt.ylabel("Number of inmigrants")
plt.show()
```

```python
ind_pak_ban = df.loc[["India","Pakistan","Bangladesh"], years]
ind_pak_ban.head()
```

```python
ind_pak_ban.T
```

```python
ind_pak_ban.T.plot()
```

### Pie plot

```python
# Total number of inmigrants for each country
continents = df.groupby("Continent", axis=0).sum()
continents
```

```python
continents["Total"].plot(kind="pie",
                         autopct="%1.1f%%",
                         shadow=True)
plt.title("Inmigration By Continents")
plt.axis("equal")
plt.show()
```

```python
continents.index
```

```python
colors = ["lightgreen", "lightblue", "pink", "purple", "royalblue", "gold"]
explode = [0.1, 0, 0, 0, 0.1, 0.1]

continents["Total"].plot(kind="pie", figsize=(17,10),
                   autopct="%1.1f%%", # display the percent value using string formatting
                   startangle=180, # start angle 90 degrees
                   shadow=True, # pie chart shadow
                   labels=None, # no labels
                   pctdistance=1.12, # percentage distance from the center of the pie
                   colors=colors, # color list
                   explode=explode) # custom explosion distance 

plt.axis("equal") # equal aspect ratio ensures that pie is drawn as a circle
plt.legend(labels=continents.index,
           loc="upper right",
           fontsize = 14)
plt.tight_layout() # adjusts subplot params so that the plot fits in figure area
plt.show()
```
